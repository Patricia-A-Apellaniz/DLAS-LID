```
# This function has been defined to scale the x matrix, as sklearn standard scaler only works for two dimensional arrays.
# It obtains the mean and the standard deviation of the matrix and the applies the (x - mean)/std formula to each of the samples
def ScaleMatrix (x):
    x_mean = np.mean(x, dtype=np.float64)
    x_std = np.std(x)
    x_result = np.zeros((len(x), len(x[0]), len(x[0,0])))
    for i in range(len(x)):
        for j in range(len(x[0])):
            for k in range(len(x[0,0])):
                x_result[i,j,k] = (x[i,j,k] - x_mean) / x_std
    print("Matrix correctly scaled")
    return x_result
```
