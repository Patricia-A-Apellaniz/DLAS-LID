```
# Needed imports
import os
import ast
import argparse
import numpy as np
import pandas as pd

from pathlib import Path
from scipy.io import savemat

import scipy.io
from sklearn.preprocessing import LabelEncoder

from google.colab import drive
drive.mount('/content/gdrive')
```

```
# Define the absolute path where data is located. 
mfcc_path = Path('/content/gdrive/My Drive/DLAS/mfccs/test')
```

```
# 3-dimensional arrays, prepared to store 431 vectors formed by 40 components in each one of their rows.
# lists in order to store the labels to predict.

# A third dimension is added (the index to identify them)
x_test = np.empty((0,40,431))
y_test = []
```

```
# Loop is designed to load the mfccs coefficients in the arrays. 
# Each row of x will have an associated value in y. 
# Loop also substitutes the language label (de, es, en) for a numerical label (0, 1, 2)

i = 0
for file in mfcc_path.iterdir():
  i=i+1
  mfccs = scipy.io.loadmat(file)
  mfccs_aux = np.array(mfccs['mfccs'], ndmin=3)
  x_test = np.concatenate((x_test, mfccs_aux))
  language = str(str(str(file).split(".")[0]).split("_")[0]).split("/")[7]
  langCode = 3;
  if language == "de":
    langCode = 0
  if language == "es":
    langCode = 1
  if language == "en":
    langCode = 2
  y_test.append(langCode)
```

```
# Convert labels to array
y_test = np.asarray(y_test)
```

```
# Save data
np.save("/content/gdrive/My Drive/DLAS/mfccs/NpyData/test/x_test.npy", x_test)
np.save("/content/gdrive/My Drive/DLAS/mfccs/NpyData/test/y_test.npy", y_test)
```
